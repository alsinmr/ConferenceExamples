```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.random.randint(2,size=[10000,10000])
```

```python
def gauss(mw,sa,A):
    x=np.linspace(mw-sa*5,mw+sa*5,1000)
    y=np.exp(-(x-mw)**2/(2*sa**2))
    y*=A/y.sum()/(x[1]-x[0])
    return x,y
```

```python
s=x.sum(0)
mw=s.mean()
sa=s.std()

ax=plt.subplots()[1]
count,bins,_=ax.hist(s,bins=50)

ax.plot(*gauss(mw,sa,count.sum()*(bins[1]-bins[0])))
```
